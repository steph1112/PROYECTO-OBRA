```python
import pandas as pd
import dash
from dash import dcc, html
from dash.dependencies import Input, Output

# Leer el archivo CSV y convertirlo en un DataFrame
df = pd.read_csv('OBRA.csv')

# Eliminar la última columna
df = df.iloc[:, :-1]

# Inicializar la aplicación Dash con external_stylesheets
external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']
app = dash.Dash(__name__, external_stylesheets=external_stylesheets)

# Layout de la aplicación
app.layout = html.Div([
    html.H1("Reporte de Tramos"),
    dcc.Dropdown(
        id='tramo-dropdown',
        options=[{'label': tramo, 'value': tramo} for tramo in df['N° DE TRAMO'].unique()],
        placeholder='Selecciona un tramo'
    ),
    html.Div(id='tramo-details')
], style={'padding': '20px', 'font-family': 'Arial'})

# Callback para actualizar los detalles del tramo seleccionado
@app.callback(
    Output('tramo-details', 'children'),
    [Input('tramo-dropdown', 'value')]
)
def update_tramo_details(selected_tramo):
    if selected_tramo:
        filtered_df = df[df['N° DE TRAMO'] == selected_tramo]
        non_empty_columns = filtered_df.columns[filtered_df.notna().any()].tolist()

        # Crear mini-tablas según las categorías especificadas
        categories = {
            'Postes': ['POSTE-15', 'POSTE-13'],
            'Conductores N2XSY': ['C-185 m2', 'C-120 m2', 'C-35 m2'],
            'Armados': ['ARM -ATVD6', 'ARM- AT5H', 'ARM- ATSB', 'ARM-ATSM', 'ARM-ATBV4', 'ARM-ATBV5', 'ARM-ATPB5'],
            'Instalacion': ['TRAFO 25KVA', 'TABLERO'],
            'Reubicacion': ['SED', 'TABLERO2'],
            'Cabezas de Terminales': ['185 m2', '120 m2', '35 m2'],
            'Equipos de Maniobra': ['Recloser', 'SBC'],
            'Retenidas': ['RO', 'RV "Y"', 'RV']
        }

        mini_tables = []
        for category, columns in categories.items():
            category_columns = [col for col in columns if col in non_empty_columns]
            if category_columns:
                mini_tables.append(html.Div([
                    html.H3(category, style={'color': '#007BFF'}),
                    html.Table([
                        html.Tr([html.Th(col, style={'background-color': '#f8f9fa', 'padding': '10px', 'border': '1px solid #dee2e6'}) for col in category_columns]),
                        html.Tr([html.Td(filtered_df.iloc[0][col], style={'padding': '10px', 'border': '1px solid #dee2e6'}) for col in category_columns])
                    ], style={'width': '100%', 'border-collapse': 'collapse', 'margin-bottom': '20px'})
                ]))

        # Tabla con las variables restantes
        remaining_columns = [col for col in non_empty_columns if not any(col in sublist for sublist in categories.values())]
        remaining_table = html.Table([
            html.Tr([html.Th(col, style={'background-color': '#f8f9fa', 'padding': '10px', 'border': '1px solid #dee2e6'}) for col in remaining_columns]),
            html.Tr([html.Td(filtered_df.iloc[0][col], style={'padding': '10px', 'border': '1px solid #dee2e6'}) for col in remaining_columns])
        ], style={'width': '100%', 'border-collapse': 'collapse', 'margin-bottom': '20px'})

        return html.Div(mini_tables + [html.H3('Otras Variables', style={'color': '#007BFF'}), remaining_table])

    return 'Selecciona un tramo para ver los detalles'

if __name__ == '__main__':
    app.run_server(debug=True)
```
